```python
# Formation of Differential Equation and Application of RK4 method to find the solution of differential Equation.
import matplotlib.pyplot as plt
import numpy as np

def f(t, y):
    x, v = y
    k = 0.1  # N/m
    m = 0.1  # kg
    dxdt = v
    dvdt = -(k / m) * x
    return np.array([dxdt, dvdt])

def RK4_step(f, t, y, dt):
    k1 = f(t, y)
    k2 = f(t + 0.5 * dt, y + 0.5 * dt * k1)
    k3 = f(t + 0.5 * dt, y + 0.5 * dt * k2)
    k4 = f(t + dt, y + dt * k3)
    return y + (dt / 6.0) * (k1 + 2 * k2 + 2 * k3 + k4)

# Initial conditions
t = 0
x0 = 0.1  # m
v0 = 0    # m/s
y0 = np.array([x0, v0])

# Time parameters
t_end = 15     # seconds 
dt = 0.01      # time step

# Arrays to store results
time_points = np.arange(0, t_end, dt)
positions = []
velocities = []

# Solve using RK4
for t in time_points:
    positions.append(y0[0])
    velocities.append(y0[1])
    y0 = RK4_step(f, t, y0, dt)

# Plotting results
plt.plot(time_points, positions, label='Position (m)')
plt.plot(time_points, velocities, label='Velocity (m/s)')
plt.xlabel('Time (s)')
plt.ylabel('Position AND Velocity')
plt.title('Simple Harmonic Motion')
plt.legend()
plt.grid(True)
plt.show()
```

### We found out that with increasing mass the time period of oscillation increases.
for mass = 0.1 kg the time of oscillation is less as compater to mass = 0.3 Kg. 


The reason for the increase in time period of oscillation with increasing mass is rooted in the fundamental principles of simple harmonic motion. According to Hooke's Law, the period of oscillation (the time it takes for one complete cycle) in a mass-spring system is inversely proportional to the square root of the mass. Mathematically, the period \( T \) is given by:

\[ $T = 2\pi \sqrt{\frac{m}{k}}$ \]

Where:
- \( T \) is the period of oscillation,
- \( m \) is the mass,
- \( k \) is the spring constant.

Since the period is inversely proportional to the square root of the mass, as the mass increases, the period also increases. This means that a heavier mass takes longer to complete one full cycle of oscillation compared to a lighter mass. Therefore, for a mass of 0.3 kg, the time period of oscillation will be longer compared to a mass of 0.1 kg.
